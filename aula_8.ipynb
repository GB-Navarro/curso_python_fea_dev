## ***Por que vamos usar bibliotecas diferentes para visualizações gráficas?***

Escolher a biblioteca certa para visualizações gráficas é crucial, já que cada uma é projetada para atender a necessidades e contextos específicos. Por isso, é fundamental conhecer as características e particularidades das diferentes bibliotecas disponíveis, compreendendo suas vantagens e limitações. A seguir, exploraremos três das mais populares: **Matplotlib**, **Seaborn** e **Plotly Express**.

- **Matplotlib**: controle total e personalização avançada
    
        Matplotlib é ideal quando você precisa de personalização detalhada e controle completo sobre seus gráficos. É uma escolha excelente para apresentações científicas ou projetos que exigem precisão e elementos customizados. Apesar de demandar mais esforço e experiência, sua flexibilidade permite criar visualizações altamente complexas, adaptadas a necessidades específicas.

- **Seaborn**: simplicidade para análise exploratória
        
        Seaborn é perfeito para análises exploratórias, especialmente quando o foco está em padrões, correlações e distribuições nos dados. Com seu design elegante e suporte nativo para gráficos estatísticos, ele permite criar visualizações esteticamente agradáveis com poucas linhas de código.

- **Plotly Express**: facilidade e interatividade

        Quando a interatividade é crucial, Plotly Express é a melhor escolha. É especialmente útil para dashboards e apresentações dinâmicas, onde os usuários precisam explorar os dados de forma autônoma. Com suporte a animações e gráficos interativos prontos para integração em aplicações web, é uma escolha popular em contextos que demandam decisões baseadas em dados, como ferramentas de business intelligence.

Em resumo, o Matplotlib é ideal para quem precisa de controle total sobre o layout e personalizações avançadas, enquanto o Seaborn se destaca por simplificar análises estatísticas, oferecendo visualizações elegantes e minimalistas. Já o Plotly Express é a melhor escolha para criar rapidamente visualizações interativas e integrá-las a dashboards.

## ***A anatomia de um gráfico***

<center><img src="plot.png"/></center>

##### ***1. Componentes da Figura***

- ***1.1 Figure:***
    
        É o contêiner de alto nível que abriga todos os elementos do gráfico. Pode conter vários eixos (subplots).

- ***1.2 Axes:***
    
        São os "eixos" onde os dados são plotados. Cada Axes pertence a uma Figure. Dentro de um Axes, temos elementos como os títulos, ticks, legendas e os próprios gráficos.

##### ***2. Elementos Principais***

- ***2.1 Title:***

        O título do gráfico, que aparece na parte superior.

- ***2.2 Legend:***
        
        A legenda que explica os elementos representados no gráfico.

- ***2.3 Grid:***

        Linhas auxiliares para facilitar a leitura dos valores.

- ***2.4 Line (line plot):***
        
        Representa uma linha conectando pontos no gráfico.

- ***2.5 Markers (scatter plot):***
        
        Marcadores que representam pontos de dados (como no scatter plot).

- ***2.6 X-axis label / Y-axis label:***

        Rótulos para os eixos X e Y, explicando o que está sendo medido.

- ***2.7 Spines:***
 
        Linhas que definem as bordas do gráfico.

##### ***3. Ticks e Labels***

- ***3.1 Major/Minor Ticks:***
        
        Marcas nos eixos para indicar divisões principais (major) e secundárias (minor).

- ***3.2 Tick Labels:***
        
        Rótulos numéricos ou textuais associados às marcas (ticks).


## ***Apresentando algumas bibliotecas para visualização gráfica usando Python***

### ***Criando um gráfico de linha usando 3 bibliotecas diferentes***

#### ***Configurando a base de dados para os próximos 3 gráficos***

```python
# Importando a biblioteca pandas para manipulação de dados.
import pandas as pd

# Criando um dicionário com dados de vendas e lucro por ano.
data = {
    "Ano": [2015, 2016, 2017, 2018, 2019, 2020, 2021],
    "Vendas": [120, 150, 170, 200, 240, 300, 350],
    "Lucro": [20, 30, 35, 50, 65, 80, 90],
}

# Convertendo o dicionário de dados para um DataFrame do pandas.
df = pd.DataFrame(data)
```

#### ***Criando um gráfico de linha usando Matplotlib***

```python
# Importa a biblioteca Matplotlib para criar o gráfico.
import matplotlib.pyplot as plt  

# Cria uma figura vazia para o gráfico.
plt.figure()

# Plota os dados diretamente no gráfico, com marcador "o", linha contínua e cor azul.
plt.plot(
    df["Ano"],  # Dados do eixo x (anos).
    df["Vendas"],  # Dados do eixo y (vendas).
    marker="o",  # Marca cada ponto com um círculo.
    linestyle="-",  # Linha contínua.
    color="blue"  # Cor azul para a linha.
)

# Adiciona o título do gráfico.
plt.title("Vendas ao longo dos anos")

# Adiciona o rótulo do eixo x (ano).
plt.xlabel("Ano")

# Adiciona o rótulo do eixo y (vendas).
plt.ylabel("Vendas")

# Salvando como imagem (A imagem deve ser salva antes de ser exibida).
plt.savefig("grafico_linha_matplotlib.png", dpi=300)  # Salva com 300 DPI para maior qualidade

# Exibe o gráfico na tela.
plt.show()

# Fecha a figura após salvar (Necessário para permitir a criação de outras figuras.)
plt.close()  
```

#### ***Criando um gráfico de linha usando Seaborn***

```python
# Importa a biblioteca Seaborn para visualização de dados
import seaborn as sns

# Cria uma figura com tamanho específico.
plt.figure()

# Plota um gráfico de linha usando o Seaborn
sns.lineplot(
    x="Ano",  # Seta os dados para o eixo x (ano).
    y="Vendas",  # Seta os dados para o eixo y (vendas).
    data=df,  # DataFrame que contém os dados a serem plotados (deve conter a coluna "ano" e a coluna "vendas").
    marker="o",  # Marca cada ponto com um círculo.
    color="green"  # Cor verde para a linha.
)

# Adiciona o título do gráfico.
plt.title("Vendas ao longo dos anos")

# Adiciona o rótulo do eixo x (ano).
plt.xlabel("Ano")

# Adiciona o rótulo do eixo y (vendas).
plt.ylabel("Vendas")

# Salvando como imagem (A imagem deve ser salva antes de ser exibida).
plt.savefig("grafico_linha_seaborn.png", dpi=300)  # Salva com 300 DPI para maior qualidade

# Exibe o gráfico na tela.
plt.show()

# Fecha a figura após salvar (Necessário para permitir a criação de outras figuras.)
plt.close()  

# Observação: O Seaborn é construído sobre o Matplotlib, por isso é comum usar métodos do Matplotlib (como plt.figure, 
# plt.title, plt.xlabel, plt.ylabel) para complementar os gráficos criados com o Seaborn.
```

#### ***Criando um gráfico de linha usando Plotly Express***

```python
# Importa a biblioteca Plotly Express para criação de gráficos interativos.
import plotly.express as px

# Cria um gráfico de linha usando o Plotly Express.
fig = px.line(
    df,  # DataFrame que contém os dados a serem plotados.
    x="Ano",  # Coluna que será usada no eixo x (ano) (Observe que o nome da coluna será o título do eixo x).
    y="Vendas",  # Coluna que será usada no eixo y (vendas) (Observe que o nome da coluna será o título do eixo y).
    title="Vendas ao longo dos anos"  # Título do gráfico.
)

# Exibe o gráfico interativo gerado
fig.show()

# Salvando o gráfico gerado como uma imagem no formato PNG (É necessário instalar o Kaleido, para tal, execute: !pip install -U Kaleido. E,
# após isso, reinicie o seu VS Code).
fig.write_image("grafico_linha_plotly_express.png")  # Também pode usar .jpg ou .pdf para outros formatos.
```

## ***Criando vários gráficos em uma mesma figure (subplots)*** 

#### ***Configurando a base de dados para os próximos 4 plots***

```python
# Vamos usar a mesma base de dados que usamos no exercício 1.

# Importando a biblioteca pandas para manipulação de dados.
import pandas as pd

# Criando um dicionário com dados de vendas, lucro, custo e margem de lucro por ano.
data = {
    "Ano": [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024],
    "Vendas": [120, 150, 170, 200, 240, 300, 350, 400, 450, 500],
    "Lucro": [20, 30, 35, 50, 65, 80, 90, 120, 150, 180],
    "Custo": [100, 120, 135, 150, 175, 220, 260, 280, 300, 320],
    "Margem_Lucro": [0.17, 0.2, 0.21, 0.25, 0.27, 0.27, 0.26, 0.3, 0.33, 0.36],
    "Despesas": [50, 60, 65, 70, 80, 100, 110, 120, 130, 140]
}

# Convertendo o dicionário de dados para um DataFrame do pandas.
df = pd.DataFrame(data)

# Exibindo o DataFrame
df
```

#### ***Usando matplotlib***

```python
import matplotlib.pyplot as plt

# Criação dos subgráficos (2 linhas x 2 colunas => 4 gráficos; com uma figure de tamanho 12 x 8).
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Gráfico 1: Vendas ao longo dos anos (Plotado na linha 0 e na coluna 0).
axs[0, 0].plot(df['Ano'], df['Vendas'], marker='o', color='blue')
axs[0, 0].set_title('Vendas ao longo dos anos')
axs[0, 0].set_xlabel('Ano')
axs[0, 0].set_ylabel('Vendas')

# Gráfico 2: Lucro ao longo dos anos (Plotado na linha 0 e na coluna 1).
axs[0, 1].plot(df['Ano'], df['Lucro'], marker='o', color='green')
axs[0, 1].set_title('Lucro ao longo dos anos')
axs[0, 1].set_xlabel('Ano')
axs[0, 1].set_ylabel('Lucro')

# Gráfico 3: Custo ao longo dos anos (Plotado na linha 1 e na coluna 0).
axs[1, 0].plot(df['Ano'], df['Custo'], marker='o', color='red')
axs[1, 0].set_title('Custo ao longo dos anos')
axs[1, 0].set_xlabel('Ano')
axs[1, 0].set_ylabel('Custo')

# Gráfico 4: Margem de lucro ao longo dos anos (Plotado na linha 1 e na coluna 1).
axs[1, 1].plot(df['Ano'], df['Margem_Lucro'], marker='o', color='purple')
axs[1, 1].set_title('Margem de Lucro ao longo dos anos')
axs[1, 1].set_xlabel('Ano')
axs[1, 1].set_ylabel('Margem de Lucro')

# Ajuste o layout para evitar sobreposição de títulos.
plt.tight_layout()

# Exibe o gráfico.
plt.show()
```

#### ***Usando plotly.express***

```python
import plotly.express as px
import plotly.subplots as sp # Para realizar subplots usando o módulo express da biblioteca Plotly é necessário usar o módulo plotly.subplots.

# Cria 4 subgráficos.
fig = sp.make_subplots(
    rows=2, cols=2, # 2 linhas x 2 colunas => 4 subgráficos.
    subplot_titles=(
        'Vendas ao longo dos anos',         # Título do subgráfico 1.
        'Lucro ao longo dos anos',          # Título do subgráfico 2.
        'Custo ao longo dos anos',          # Título do subgráfico 3.
        'Margem de Lucro ao longo dos anos' # Título do subgráfico 4.
    )
)

# Gráfico 1: Vendas ao longo dos anos.
fig.add_trace(
    px.line(df, x='Ano', y='Vendas').data[0], # .data[0] precisa ser usado devido ao tipo de retorno do método px.line.
    row=1, col=1
)

# Gráfico 2: Lucro ao longo dos anos.
fig.add_trace(
    px.line(df, x='Ano', y='Lucro').data[0], # .data[0] precisa ser usado devido ao tipo de retorno do método px.line.
    row=1, col=2
)

# Gráfico 3: Custo ao longo dos anos.
fig.add_trace(
    px.line(df, x='Ano', y='Custo').data[0], # .data[0] precisa ser usado devido ao tipo de retorno do método px.line.
    row=2, col=1
)

# Gráfico 4: Margem de Lucro ao longo dos anos.
fig.add_trace(
    px.line(df, x='Ano', y='Margem_Lucro').data[0], # .data[0] precisa ser usado devido ao tipo de retorno do método px.line.
    row=2, col=2
)

# Atualiza o layout para colocar o título do plot (conjunto de gráficos).
fig.update_layout(
    title_text="Análise de Vendas, Lucro, Custo e Margem de Lucro ao longo dos anos", 
    showlegend=False
)

# Exibe o gráfico.
fig.show()
```

#### ***Usando plotly.graph_objects***

*O módulo plotly.graph_objects é, na minha opinião, uma das melhores opções de visualização gráfica no Python. Ele combina a interatividade do plotly.express com a flexibilidade e a alta customização do matplotlib. Embora a curva de aprendizado desse módulo seja um pouco mais elevada, ele oferece um nível de controle e personalização que pode ser muito vantajoso, especialmente para visualizações mais complexas.*

*Abaixo está um exemplo de como criar gráficos usando o plotly.graph_objects — essa abordagem será útil para quem deseja um controle mais detalhado sobre a criação e o estilo dos gráficos. Atenção: este exemplo é fornecido como uma sugestão para estudo futuro e não será cobrado nessa aula.*

```python
import plotly.graph_objects as go
import plotly.subplots as sp # Para realizar subplots usando o módulo graph_objects da biblioteca Plotly também é necessário  
                             # usar o módulo plotly.subplots.

# Cria 4 subgráficos.
fig = sp.make_subplots(
    rows=2, cols=2, # 2 linhas x 2 colunas => 4 subgráficos.
    subplot_titles=(
        'Vendas ao longo dos anos',         # Título do subgráfico 1.
        'Lucro ao longo dos anos',          # Título do subgráfico 2.
        'Custo ao longo dos anos',          # Título do subgráfico 3.
        'Margem de Lucro ao longo dos anos' # Título do subgráfico 4.
    )
)

# Gráfico 1: Vendas ao longo dos anos.
fig.add_trace(
    go.Scatter(x=df['Ano'], y=df['Vendas'], mode='lines+markers', name='Vendas'),
    row=1, col=1
)

# Gráfico 2: Lucro ao longo dos anos.
fig.add_trace(
    go.Scatter(x=df['Ano'], y=df['Lucro'], mode='lines+markers', name='Lucro'),
    row=1, col=2
)

# Gráfico 3: Custo ao longo dos anos.
fig.add_trace(
    go.Scatter(x=df['Ano'], y=df['Custo'], mode='lines+markers', name='Custo'),
    row=2, col=1
)

# Gráfico 4: Margem de Lucro ao longo dos anos.
fig.add_trace(
    go.Scatter(x=df['Ano'], y=df['Margem_Lucro'], mode='lines+markers', name='Margem de Lucro'),
    row=2, col=2
)

# Atualiza o layout para colocar o título do plot (conjunto de gráficos).
fig.update_layout(
    title_text="Análise de Vendas, Lucro, Custo e Margem de Lucro ao longo dos anos", 
    showlegend=False
)

# Exibe o gráfico.
fig.show()
```

#### ***Multiplos gráficos de linha em um mesmo plot***

```python
# Agrupa as vendas totais por produto.
df_total_vendas = df_vendas.groupby("Produto")["Vendas"].sum().reset_index()  

# Cria o gráfico de pizza.
fig_pie = px.pie(
    df_total_vendas,  # DataFrame contendo os dados agregados.
    values="Vendas",  # Valores das fatias (vendas totais por produto).
    names="Produto",  # Rótulos das fatias (nome do produto).
    title="Distribuição de Vendas por Produto",  # Define o título do gráfico.
    hole=0.4  # Adiciona um buraco ao centro, transformando o gráfico em um gráfico de "rosca".
)

# Personaliza o gráfico para mostrar o rótulo e o valor total ao passar o mouse sobre ele.
fig_pie.update_traces(
    hoverinfo="label+value"  
)

# Exibe o gráfico gerado.
fig_pie.show()
```
